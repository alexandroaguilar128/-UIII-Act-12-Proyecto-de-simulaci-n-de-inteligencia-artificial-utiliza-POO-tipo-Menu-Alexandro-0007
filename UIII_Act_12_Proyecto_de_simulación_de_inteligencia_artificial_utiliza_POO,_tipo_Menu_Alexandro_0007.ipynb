### Medico

```python
class Medico:
    def __init__(self, nombre, lic, esp, hosp, exp, pac, turno):
        # Atributos de la instancia (Estado del objeto) [6, 7]
        self.nombre = nombre
        self._licencia = lic
        self._especialidad = esp
        self._hospital = hosp
        self._años_exp = exp
        self._pacientes = pac
        self._turno = turno

    def licencia_profesional(self):
        return self._licencia

    def especialidad(self):
        return self._especialidad

    def hospital_adscripcion(self):
        return self._hospital

    def años_experiencia(self):
        return self._años_exp

    def pacientes_atendidos(self):
        return self._pacientes

    def turno(self):
        return self._turno

    def recomendacion_ia(self, sintoma):
        print(f"\nIA analizando perfil del Dr(a). {self.nombre}...")
        if sintoma == "fiebre" and self._especialidad == "general":
            return "RESULTADO IA: Perfil apto para consulta inmediata."
        elif sintoma == "cirugia" and self._años_exp > 10:
            return "RESULTADO IA: Especialista con alta trayectoria para el caso."
        else:
            return "RESULTADO IA: Los datos no coinciden plenamente. Consulte otro perfil."

# --- Lógica del Menú Interactivo ---
lista_medicos = []

print("--- SISTEMA DE INTELIGENCIA MÉDICA v6.0 (POO) ---")

while True:
    print("\n--- MENÚ DE CONTROL ---")
    print("1. Registrar nuevo médico (Instanciar)")
    print("2. Ver listado de médicos (Tabla alineada)")
    print("3. Consultar recomendación de la IA")
    print("4. Salir")
    opcion = input("Seleccione una opción (1-4): ")

    if opcion == "1":
        print("\n--- Entrada de Datos (Key, Values) ---")
        n = input("Nombre completo: ")
        l = input("Licencia profesional: ")
        e = input("Especialidad (general/cirujano): ").lower()
        h = input("Hospital de adscripción: ")
        exp = int(input("Años de experiencia: "))
        p = int(input("Pacientes atendidos al mes: "))
        t = input("Turno (matutino/vespertino): ")

        # Creación del objeto (Instanciación) [11, 12]
        nuevo_medico = Medico(n, l, e, h, exp, p, t)
        lista_medicos.append(nuevo_medico)
        print("SISTEMA: Médico registrado y encapsulado correctamente.")

    elif opcion == "2":
        if not lista_medicos:
            print("IA AVISO: El directorio está vacío.")
        else:
            print("\n" + "="*115)
            # Tabla bien alineada usando formato de cadenas [1264, 18.10.1]
            print("{:<15} | {:<12} | {:<12} | {:<15} | {:<8} | {:<10} | {:<10}".format(
                "NOMBRE", "LICENCIA", "ESPECIALIDAD", "HOSPITAL", "EXP", "PACIENTES", "TURNO"))
            print("-" * 115)
            for m in lista_medicos:
                # Se llaman a las funciones solicitadas para obtener la información [13]
                print("{:<15} | {:<12} | {:<12} | {:<15} | {:<8} | {:<10} | {:<10}".format(
                    m.nombre, m.licencia_profesional(), m.especialidad(),
                    m.hospital_adscripcion(), m.años_experiencia(),
                    m.pacientes_atendidos(), m.turno()))
            print("="*115)

    elif opcion == "3":
        if not lista_medicos:
            print("IA ERROR: No hay médicos registrados para consultar.")
        else:
            sintoma = input("Describa la necesidad (fiebre/cirugia): ").lower()
            # La IA evalúa al primer médico de la lista como ejemplo
            print(lista_medicos.recomendacion_ia(sintoma))

    elif opcion == "4":
        print("IA: Apagando servidores médicos. ¡Adiós!")
        break # Finaliza el ciclo [4.4, 64]
print("Elaborado por Alexandro Aguilar NL = 1")
```
