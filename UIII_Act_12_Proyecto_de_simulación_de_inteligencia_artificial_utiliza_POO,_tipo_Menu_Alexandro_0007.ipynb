### Medico

```python
# 1. Definimos la Clase con sus atributos
class Medico:
    def __init__(self, licencia, especialidad, hospital, experiencia, pacientes, turno):
        self.licencia = licencia
        self.especialidad = especialidad
        self.hospital = hospital
        self.experiencia = experiencia
        self.pacientes = pacientes
        self.turno = turno

def agregar_medico():
    licencia = input("Licencia: ")
    especialidad = input("Especialidad: ")
    hospital = input("Hospital: ")
    experiencia = input("Años experiencia: ")
    pacientes = input("Pacientes atendidos: ")
    turno = input("Turno: ")

    nuevo_medico = Medico(licencia, especialidad, hospital, experiencia, pacientes, turno)
    medicos.append(nuevo_medico)
    print("¡Médico agregado!")

def listar_medicos():
    if not medicos:
        print("No hay médicos registrados.")
        return

def actualizar_medico():
    busca = input("Licencia del médico a modificar: ")
    for m in medicos:
        if m.licencia == busca:
            m.especialidad = input("Nueva especialidad: ")
            m.hospital = input("Nuevo hospital: ")
            m.experiencia = input("Nuevos años experiencia: ")
            m.pacientes = input("Nuevos pacientes atendidos: ")
            m.turno = input("Nuevo turno: ")
            print("¡Médico actualizado!")
            return
    print("No se encontró esa licencia.")

def borrar_medico():
    busca = input("Licencia del médico a borrar: ")
    for m in medicos:
        if m.licencia == busca:
            medicos.remove(m)
            print("¡Médico borrado!")
            return
    print("No se encontró esa licencia.")


  # Encabezados con formato y bordes de doble línea
    print(f"\n{'Licencia':<12} ║ {'Especialidad':<15} ║ {'Hospital':<15} ║ {'Exp (Años)':<12} ║ {'Pacientes':<10} ║ {'Turno':<10}")
    print("═" * 87) # Línea divisoria doble y más gruesa

  # Datos de los médicos separados por los bordes gruesos
    for m in medicos:
        print(f"{m.licencia:<12} ║ {m.especialidad:<15} ║ {m.hospital:<15} ║ {m.experiencia:<12} ║ {m.pacientes:<10} ║ {m.turno:<10}")

# Lista donde guardaremos los objetos de tipo Medico
medicos = []

# 3. Menú interactivo
while True:
    print("\n1. Agregar | 2. Listar | 3. Actualizar | 4. Borrar | 5. Salir")
    opcion = input("Opción: ")

    if opcion == "1": agregar_medico()
    elif opcion == "2": listar_medicos()
    elif opcion == "3": actualizar_medico()
    elif opcion == "4": borrar_medico()
    elif opcion == "5":
        print("¡Adiós!")
        break
print("Elaborado por Alexandro Aguilar NL = 1")
```
