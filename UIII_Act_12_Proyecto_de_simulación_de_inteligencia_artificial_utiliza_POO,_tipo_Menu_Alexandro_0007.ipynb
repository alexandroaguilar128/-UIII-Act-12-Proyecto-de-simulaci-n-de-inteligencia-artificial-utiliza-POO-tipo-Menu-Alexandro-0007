### Medico

```python
class Medico:
    # Atributos de la instancia (Constructor) [7, 8]
    def __init__(self, nombre, especialidad, experiencia, pacientes):
        self.nombre = nombre
        self.especialidad = especialidad
        self.experiencia = experiencia
        self.pacientes = pacientes

    # Función dentro de la clase para simular el diagnóstico IA [9, 10]
    def realizar_diagnostico_ia(self, sintoma, temperatura):
        print(f"\nIA analizando con el perfil del Dr(a). {self.nombre}...")

        # Lógica de decisión de la IA basada en el conocimiento del médico [11, 12]
        if temperatura > 38.0 and sintoma == "tos":
            return "Resultado: Infección respiratoria. Acción: Reposo y aislamiento."
        elif temperatura > 37.0 and sintoma == "dolor":
            return "Resultado: Cuadro inflamatorio. Acción: Hidratación y monitoreo."
        else:
            return "Resultado: Signos estables. Acción: Mantener observación."

    # Función para dar formato a la fila de la tabla [4, 13]
    def obtener_fila_tabla(self):
        return "{:<15} | {:<15} | {:<10} | {:<10}".format(
            self.nombre, self.especialidad, self.experiencia, self.pacientes
        )

# --- Lógica del Programa Principal ---

directorio_medicos = [] # Lista para almacenar objetos Medico [14]

print("--- BIENVENIDO AL SISTEMA MÉDICO IA (MODO POO) ---")

while True:
    print("\nMENÚ PRINCIPAL:")
    print("1. Registrar nuevo médico")
    print("2. Ver directorio de médicos (Tabla)")
    print("3. Iniciar consulta inteligente (IA)")
    print("4. Salir")

    opcion = input("Seleccione una opción (1-4): ")

    # CASO 1: Entrada de datos y creación de objetos [15, 16]
    if opcion == "1":
        print("\n--- Registro de Datos ---")
        nombre = input("Nombre: ")
        especialidad = input("Especialidad: ")
        experiencia = int(input("Años de experiencia: "))
        pacientes = int(input("Pacientes actuales: "))

        # Instanciación de la clase: creamos el objeto médico [17, 18]
        nuevo_medico = Medico(nombre, especialidad, experiencia, pacientes)
        directorio_medicos.append(nuevo_medico)
        print("SISTEMA: Médico instanciado y guardado correctamente.")

    # CASO 2: Listado en forma de tabla alineada [19, 20]
    elif opcion == "2":
        if not directorio_medicos:
            print("SISTEMA: No hay médicos registrados.")
        else:
            print("\n" + "="*60)
            print("{:<15} | {:<15} | {:<10} | {:<10}".format("NOMBRE", "ESPECIALIDAD", "EXP", "PACIENTES"))
            print("-" * 60)
            # Recorremos la lista de objetos con un ciclo for [20, 21]
            for m in directorio_medicos:
                print(m.obtener_fila_tabla())
            print("="*60)

    # CASO 3: Interacción con la IA [22, 23]
    elif opcion == "3":
        if not directorio_medicos:
            print("IA ERROR: Debe registrar al menos un médico primero.")
        else:
            sintoma = input("IA: Describa el síntoma (tos / dolor / ninguno): ").lower()
            temperatura = float(input("IA: Ingrese temperatura (°C): "))

            # Usamos el primer médico disponible para procesar el diagnóstico [24, 25]
            medico_ia = directorio_medicos
            resultado = medico_ia.realizar_diagnostico_ia(sintoma, temperatura)
            print(resultado)

    # CASO 4: Salida [26, 27]
    elif opcion == "4":
        print("IA: Cerrando sesión. Los objetos han sido eliminados de la memoria.")
        break # Rompe el ciclo para finalizar el programa [28, 29]
print("Elaborado por Alexandro Aguilar NL }= 1")
```
