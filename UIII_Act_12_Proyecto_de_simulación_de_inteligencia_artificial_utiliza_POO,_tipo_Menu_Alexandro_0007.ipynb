```python
# 1. Definimos la Clase
class Medico:
    # Las funciones se pasan aquí adentro de los paréntesis de __init__
    def __init__(self, agregar_func, listar_func, actualizar_func, borrar_func):
        # Guardamos las funciones dentro del objeto
        self.agregar = agregar_func
        self.listar = listar_func
        self.actualizar = actualizar_func
        self.borrar = borrar_func

    def ejecutar_menu(self):
        while True:
            print("\n1. Agregar | 2. Listar | 3. Actualizar | 4. Borrar | 5. Salir")
            opcion = input("Opción: ")

            # Ejecutamos las funciones que se pasaron por los paréntesis
            if opcion == "1": self.agregar()
            elif opcion == "2": self.listar()
            elif opcion == "3": self.actualizar()
            elif opcion == "4": self.borrar()
            elif opcion == "5":
                print("¡Adiós!")
                break

def agregar_medico():
    licencia = input("Licencia: ")
    especialidad = input("Especialidad: ")
    hospital = input("Hospital: ")
    experience = input("Años experiencia: ")
    pacientes = input("Pacientes atendidos: ")
    turno = input("Turno: ")

    nuevo_medico = DatosMedico(licencia, especialidad, hospital, experience, pacientes, turno)
    medicos.append(nuevo_medico)
    print("¡Médico agregado!")

def listar_medicos():
    if not medicos:
        print("No hay médicos registrados.")
        return
    print(f"\n{'Licencia':<12} ║ {'Especialidad':<15} ║ {'Hospital':<15} ║ {'Exp (Años)':<12} ║ {'Pacientes':<10} ║ {'Turno':<10}")
    print("═" * 87)
    for m in medicos:
        print(f"{m.licencia:<12} ║ {m.especialidad:<15} ║ {m.hospital:<15} ║ {m.experiencia:<12} ║ {m.pacientes:<10} ║ {m.turno:<10}")

def actualizar_medico():
    busca = input("Licencia del médico a modificar: ")
    for m in medicos:
        if m.licencia == busca:
            m.especialidad = input("Nueva especialidad: ")
            m.hospital = input("Nuevo hospital: ")
            m.experiencia = input("Nuevos años experiencia: ")
            m.pacientes = input("Nuevos pacientes atendidos: ")
            m.turno = input("Nuevo turno: ")
            print("¡Médico actualizado!")
            return
    print("No se encontró esa licencia.")

def borrar_medico():
    busca = input("Licencia del médico a borrar: ")
    for m in medicos:
        if m.licencia == busca:
            medicos.remove(m)
            print("¡Médico borrado!")
            return
    print("No se encontró esa licencia.")

print("\nElaborado por Alexandro Aguilar NL = 1")

# 2. Datos y Funciones de la aplicación (se mantienen igual a tu código original)
class DatosMedico:
    def __init__(self, licencia, especialidad, hospital, experiencia, pacientes, turno):
        self.licencia = licencia
        self.especialidad = especialidad
        self.hospital = hospital
        self.experiencia = experiencia
        self.pacientes = pacientes
        self.turno = turno

medicos = []

sistema = Medico(agregar_medico, listar_medicos, actualizar_medico, borrar_medico)
sistema.ejecutar_menu()
```
